## replace , with |

```python
import pandas as pd
import os
csv_folder = "/content/csv"
for filename in os.listdir(csv_folder):
    if filename.endswith(".csv"):
      filename_path = os.path.join(csv_folder, filename)
      filename_outpath = os.path.join("/content/out", filename)
      pd.read_csv(filename_path).assign(lake_ids=lambda df: df["lake_ids"].astype(str).str.replace(",","|")).to_csv(filename_outpath, index=False)
```

# doublons

```python
import pandas as pd
import os

csv_brut_folder = "/content"

for filename in os.listdir(csv_brut_folder):
    if filename.endswith(".csv"):
        filename_path = os.path.join(csv_brut_folder, filename)
        df = pd.read_csv(filename_path)

        # Identifier les doublons sur latitude + longitude
        duplicates = df[df.duplicated(subset=["latitude", "longitude"], keep=False)]

        if not duplicates.empty:
            print(f"❗ Doublons trouvés dans le fichier : {filename}")
            print(duplicates)
            print("\nNombre de doublons :", len(duplicates))
            print("=" * 80)
        else:
            print(f"✅ Aucun doublon trouvé dans le fichier : {filename}")
```

```python
import pandas as pd
import os

csv_brut_folder = "/content"

for filename in os.listdir(csv_brut_folder):
    if filename.endswith(".csv"):
        filename_path = os.path.join(csv_brut_folder, filename)
        df = pd.read_csv(filename_path)

        # Trouver les lignes ayant des doublons sur latitude + longitude
        dup_mask = df.duplicated(subset=["latitude", "longitude"], keep=False)
        duplicates = df[dup_mask].copy()

        if not duplicates.empty:
            print(f"\n❗ Doublons trouvés dans le fichier : {filename}")

            # Trier pour faciliter la lecture
            duplicates = duplicates.sort_values(by=["latitude", "longitude", "elevation"])

            # Grouper par latitude + longitude
            grouped = duplicates.groupby(["latitude", "longitude"])

            for (lat, lon), group in grouped:
                print(f"\n--- Latitude: {lat}, Longitude: {lon} (x{len(group)}) ---")
                print(group.reset_index(drop=True))
        else:
            print(f"✅ Aucun doublon trouvé dans le fichier : {filename}")
```

# suppression doublons puis pivot

```python
import pandas as pd
import os

csv_brut_folder = "/content"

for filename in os.listdir(csv_brut_folder):
    if filename.endswith(".csv"):
        filename_path = os.path.join(csv_brut_folder, filename)
        df = pd.read_csv(filename_path)
        # supprimer les doublons
        df = df.drop_duplicates(subset=["longitude", "latitude", "elevation"])

        # Créer le pivot : index = latitude, colonnes = longitude, valeurs = ????????????
        pivoted = df.pivot(index="latitude", columns="longitude", values="elevation")

        # Trier l'index (latitude) dans le bon ordre (souvent décroissant pour imager du nord vers le sud)
        pivoted = pivoted.sort_index(ascending=False)

        # Sauvegarder le fichier pivoté
        filename_outpath = os.path.join("/content", filename.replace(".csv", "_pivoted.csv"))
        pivoted.to_csv(filename_outpath)
```
