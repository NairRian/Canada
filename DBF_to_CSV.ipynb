```python
import pandas as pd
import os
import re

path_zones = "/content/drive/MyDrive/Colab_Notebooks/Canada/Lakes"   # dossier avec Lake_attributes_zone_XX.csv
lake_global = "/content/drive/MyDrive/Colab_Notebooks/Canada/LakeTEMP_aggregated_v1.csv"  # fichier global
output_folder = "/content/drive/MyDrive/Colab_Notebooks/Canada/Lakes_merged"

os.makedirs(output_folder, exist_ok=True)

# Charger le fichier global une seule fois
df_global = pd.read_csv(lake_global)

for file in os.listdir(path_zones):
    # On ne traite que les bons fichiers
    match = re.search(r"Lake_attributes_zone_(\d+)\.csv", file)
    if not match:
        print(f"❌ Fichier ignoré (pas un Lake_attributes_zone_XX.csv) : {file}")
        continue

    zone = int(match.group(1))
    zone_file = os.path.join(path_zones, file)
    df_zone = pd.read_csv(zone_file)

    # Jointure sur Hylak_id
    df_merge = df_zone.merge(
        df_global,
        on="Hylak_id",
        how="left",
        suffixes=("", "_drop")   # les colonnes du global auront le suffixe "_drop"
    )

    # Supprimer les colonnes doublons (celles finissant par "_drop")
    df_merge = df_merge.loc[:, ~df_merge.columns.str.endswith('_drop')]

    # Sauvegarde
    output_file = os.path.join(output_folder, f"Lake_zone_{zone}.csv")
    df_merge.to_csv(output_file, index=False, encoding="utf-8")
    print(f"✅ {file} → {os.path.basename(output_file)}")

print("\n📂 Fusion terminée ! Les fichiers CSV enrichis sont dans :", output_folder)
```

```python
import pandas as pd
import os
import re

path_HR = "/content/drive/MyDrive/Colab_Notebooks/Canada/HydroRivers"
path_RA = "/content/drive/MyDrive/Colab_Notebooks/Canada/RiverAtlas"
output_folder = "/content/drive/MyDrive/Colab_Notebooks/Canada/CSV"

for file in os.listdir(path_HR):

    # Extract the zone's number from the name of the file
    match = re.search(r"Rivers_zone_(\d+)", file) # rename the search depending on the filename format
    if match:
        zone = int(match.group(1))
    else:
        print(f"❌ Impossible to find the zone's number in this file : {file}")
        continue

    river_file = os.path.join(path_RA, f"Rivers_attributes_zone_{zone}.csv")
    if not os.path.exists(river_file):
        print(f"⚠️ Pas de RiverAtlas pour la zone {zone}")
        continue

    # Charger les deux tables
    hydro_file = os.path.join(path_HR, file)
    df_hydro = pd.read_csv(hydro_file)
    df_river = pd.read_csv(river_file)

    # Jointure sur l’ID unique (exemple HYRIV_ID)
    df = df_river.merge(df_hydro[["HYRIV_ID","X_start","Y_start","X_end","Y_end"]],
                    on="HYRIV_ID", how="left")

    output_file = os.path.join(output_folder, f"River_zone_{zone}.csv")
    df.to_csv(output_file, index=False, encoding="utf-8")
    print(f"✅ {file} → {os.path.basename(output_file)}")

print("\n📂 Conversion terminée ! Les fichiers CSV sont dans :", output_folder)
```

```python
import os
from dbfread import DBF
import pandas as pd
```

```python
# --- Paramètres ---
# Dossier contenant les fichiers .dbf
# Dans Colab, tu peux uploader un dossier depuis ton ordi ou utiliser Google Drive
dbf_folder = "/content/DBF"
output_folder = "/content/drive/MyDrive/Colab_Notebooks/Canada/CSV"

# --- Boucle sur tous les fichiers .dbf ---
for file in os.listdir(dbf_folder):
    if file.lower().endswith(".dbf"):
        dbf_path = os.path.join(dbf_folder, file)
        csv_path = os.path.join(output_folder, os.path.splitext(file)[0] + ".csv")

        try:
            # Lecture du DBF avec encodage CP1252 (classique en France)
            table = DBF(dbf_path, encoding="cp1252")
            df = pd.DataFrame(iter(table))

            # Export CSV UTF-8
            df.to_csv(csv_path, index=False, encoding="utf-8")
            print(f"✅ {file} → {os.path.basename(csv_path)}")

        except Exception as e:
            print(f"❌ Erreur sur {file} : {e}")

print("\n📂 Conversion terminée ! Les fichiers CSV sont dans :", output_folder)
```

```python
import os

# Nom du dossier principal
base_dir = "/content/drive/MyDrive/Colab_Notebooks/Canada/Results"

# Crée le dossier "results" s'il n'existe pas
os.makedirs(base_dir, exist_ok=True)

# Boucle sur les dossiers 1 à 9
for i in range(1, 10):
    folder_i = os.path.join(base_dir, str(i))
    os.makedirs(folder_i, exist_ok=True)

    # Boucle pour les sous-dossiers, ex : 11 à 19 pour i=1
    for j in range(1, 10):
        sub_number = int(f"{i}{j}")  # ex : 1 et 1 -> 11
        sub_folder = os.path.join(folder_i, str(sub_number))
        os.makedirs(sub_folder, exist_ok=True)

        # Crée les deux sous-sous-dossiers
        os.makedirs(os.path.join(sub_folder, f"{sub_number}_flow_accumulation"), exist_ok=True)
        os.makedirs(os.path.join(sub_folder, f"{sub_number}_flow_direction"), exist_ok=True)
        os.makedirs(os.path.join(sub_folder, f"{sub_number}_elevation_avg"), exist_ok=True)
        os.makedirs(os.path.join(sub_folder, f"{sub_number}_lakes_ID_list"), exist_ok=True)
        os.makedirs(os.path.join(sub_folder, f"{sub_number}_land_mask"), exist_ok=True)
        os.makedirs(os.path.join(sub_folder, f"{sub_number}_rivers"), exist_ok=True)
        os.makedirs(os.path.join(sub_folder, f"{sub_number}_watershed_ID"), exist_ok=True)

print("Structure de dossiers créée avec succès ✅")
```
